# Seaborn

Seaborn is a Python data visualization library based on matplotlib.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
'''
Other dataset present
1. Tips Dataset
2. Iris Dataset
3. Penguins Dataset
4. Flights Dataset
5. Diamonds Dataset
6. Titanic Dataset
7. Exercise Dataset
8. MPG Dataset
9. Planets Dataset
'''
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips.head()
```

```python
# HISTPLOT
sns.histplot(data = tips, x='total_bill', bins=20, kde=True)
plt.title("Histogram of Total Bill with KDE")
plt.show()
```

```python
# KDEPLOT
sns.kdeplot(data=tips['tip'], fill=True, bw_adjust=0.5)
plt.title("KDE Plot of Tip Amount")
plt.show()
```

```python
# NEW DATASET
iris = sns.load_dataset('iris')
iris.head()
```

```python
# DISPLOT
sns.displot(
    data=iris,
    x='sepal_length',
    hue='species',
    kind='kde',
    fill=True
)

plt.title("Sepal Length Distribution by species")
plt.show()
```

```python
# BARPLOT
sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    ci='sd'
)
plt.title('Average total bill by day and gender')
plt.show()
```

```python
# BOXPLOT
sns.boxplot(
    data=tips,
    x='day',
    y='tip',
    hue='sex'
)
plt.title('Boxplot of tip by day and gender')
plt.show()
```

```python
# VIOLINPLOT
sns.violinplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    split=True
)
plt.title('Violin plot of total bill')
plt.show()
```

```python
# STRIPPLOT
sns.stripplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    jitter=True,
    dodge=True
)
plt.title('Raw tip data by day and gender')
plt.show()
```

```python
# SWARMPLOT
sns.swarmplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    dodge=True
)
plt.title('Swarmplot of bills')
plt.show()
```

```python
# REGPLOT
sns.regplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title('Total bill vs Tip with regression line')
plt.show()
```

```python
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='smoker'
)
plt.title('LM Plot, Gender & smoker wise Trend', y = 1.1)
plt.show()
```

```python
sns.pairplot(
    data=iris,
    hue='species'
)
plt.show()
```
